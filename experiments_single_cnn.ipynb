```python
import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
import birdloader
from birdloader import BirdImage as BirdImage
import function as fn
from torch.optim.lr_scheduler import CosineAnnealingLR
torch.manual_seed(0)
```

```python
BATCHSIZE = 16

trainset = BirdImage('train')
devset = BirdImage('dev')

train_loader = DataLoader(dataset = trainset,
                          batch_size = BATCHSIZE,
                          shuffle = True,
                          num_workers = 0,
                          pin_memory = True)

dev_loader = DataLoader(dataset = devset,
                        batch_size = BATCHSIZE,
                        shuffle = False,
                        num_workers = 0,
                        pin_memory = True)
```

```python
# model1 = models.resnet152(pretrained=True)
# for param in model1.parameters():
#     param.requires_grad = False
#model1.fc = nn.Linear(2048, 200)
model = models.densenet121(pretrained=True)
model.classifier = nn.Linear(1024, 200)
info = {
    'model':model, 
    'train_loader':train_loader, 
    'dev_loader':dev_loader, 
    'optimizer':None,
    'scheduler':None,
    'criterion':None,

    'EPOCHS':60, 
    'MODELNAME':'densenet121'
}
info['criterion'] = nn.CosineSimilarity()
info['optimizer'] = torch.optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=0.001)
info['scheduler'] = optim.lr_scheduler.CosineAnnealingLR(info['optimizer'], T_max=6, verbose=True)
fn.train_new(**info)
# fn.train(**info)
```

```python
model = models.resnet152(pretrained=True)
model.fc = nn.Linear(2048, 200)

info2 = {
    'model':model, 
    'train_loader':train_loader, 
    'dev_loader':dev_loader, 
    'optimizer':None,
    'scheduler':None,
    'criterion':None,

    'EPOCHS':60, 
    'MODELNAME':'resnet'
}
info['criterion'] = nn.CosineSimilarity()
info['optimizer'] = torch.optim.SGD(filter(lambda p: p.requires_grad, model.parameters()), lr=0.001, weight_decay=5e-4)
info['scheduler'] = optim.lr_scheduler.CosineAnnealingLR(info['optimizer'], T_max=6, verbose=True)
fn.train_new(**info2)
# fn.train(**info2)
```
