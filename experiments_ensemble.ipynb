```python
'''
this part is do feature extraction from three main nets
first is resnet
second is vggnet
third is googlenet
first of all, we remove all the fc layers in these three nets
and do feature extraction via these three nets
second
concatnate all features extracted from nets
and put it into the MLP layer to do classification
'''
SIZE = 4096
class Resnet2equalsize(nn.Module):
    '''
    input:[-1, 2048, 1, 1] tensor
    output:[-1, SIZE]
    '''
    def __init__(self, size):
        super().__init__()
        self.size = size
        self.convert = nn.Sequential(
            nn.Flatten(),
            nn.Linear(2048, self.size)
        )
    def forward(self, x):
        return self.convert(x)
    
class Vggnet2equalsize(nn.Module):
    '''
    input:[-1, 4096] tensor(classifier first linear layer)
    output:[-1, SIZE]
    '''
    def __init__(self):
        super().__init__()
        self.convert = nn.Sequential(
            nn.Flatten()# we dont need to add a new linear layer
        )
    def forward(self, x):
        return self.convert(x)
    
class Googlenet2equalsize(nn.Module):
    '''
    input:[-1, 1024, 1, 1] tensor
    output:[-1, SIZE]
    '''
    def __init__(self, size):
        super().__init__()
        self.size = size
        self.convert = nn.Sequential(
            nn.Flatten(),
            nn.Linear(1024, self.size)
        )
    def forward(self, x):
        return self.convert(x)

class FusionNet(nn.Module):
    def __init__(self, size, r2es, v2es, g2es):
        super().__init__()
        
        self.size = size
        
        self.r2es = r2es
        self.v2es = v2es
        self.g2es = g2es
        
        self.net = nn.Sequential(
            nn.Linear(self.size * 3, self.size),
            nn.ReLU(),
            nn.Dropout(0.7),
            
            nn.Linear(self.size, 1024),
            nn.ReLU(),
            nn.Dropout(0.7),
            
            nn.Linear(1024, 200)
        )
        
    def forward(self, x1, x2, x3):
        x1 = self.r2es(x1)#[-1, size]
        print(x1.shape)
        x2 = self.v2es(x2)#[-1, size]
        print(x2.shape)
        x3 = self.g2es(x3)#[-1, size]
        print(x3.shape)
        x = torch.cat([x1, x2, x3], dim = 1)
        
        x = self.net(x)
        return x
        
```

```python
R2ES = Resnet2equalsize(SIZE)
V2ES = Vggnet2equalsize()
G2ES = Googlenet2equalsize(SIZE)

fmodel = FusionNet(SIZE, R2ES, V2ES, G2ES)
```
